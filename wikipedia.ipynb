```python
from bs4 import BeautifulSoup
import requests
```

PROGRAM TO DISPLAY HEADER FROM WIKIPEDIA

```python
page=requests.get("https://en.wikipedia.org/wiki/Main_Page")
page
```

we can observe the above data is unstructured

```python
Soup=BeautifulSoup(page.content,'html.parser')
Soup
```

below code to select all items which has headline

```python
head=Soup.find_all('span',class_="mw-headline")
head
```

using for loop to extract all the headings in the webpage

```python
heading=[]
for i in head:
    heading.append(i.text)
```

```python
heading
```

```python
heading=[]
for i in head:
    heading.append(i.text.replace('\xa0...',''))
```
